```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
# 학적 데이터 불러오기
filename = '/content/drive/My Drive/Major/1130/student.csv'
student = pd.read_csv(filename, encoding='utf-8')
student = student.drop(student.columns[1],axis=1)
student = student.drop(student.columns[1],axis=1)
student = student.drop(student.columns[1],axis=1)
student = student.drop(student.columns[2],axis=1)
student = student.drop(student.columns[2],axis=1)
student.head(50)
```

```python
# 수업 데이터 불러오기
filename = '/content/drive/My Drive/Major/1130/take.csv'
take = pd.read_csv(filename, encoding='utf-8')
take = take.drop(take.columns[2],axis=1)
take = take.drop(take.columns[2],axis=1)
take = take.drop(take.columns[2],axis=1)
take = take.drop(take.columns[2],axis=1)
take = take.drop(take.columns[2],axis=1)
take.head(50)
```

```python
MergeTable = pd.merge(student, take, how='inner', on=None)
MergeTable.head(50)
```

```python
# 해당 학년도 제외 삭제
MergeTable = MergeTable[MergeTable['입학학년도']==MergeTable['학년도']]

MergeTable.head(50)
```

```python
# ID 기준 그룹화
idGroup = MergeTable.groupby('ID').학점.sum().reset_index()
idGroup.head(50)
```

```python
# 이수학점 점수 추가
idGroup['takepoint'] = idGroup['학점']/36*150

idGroup.loc[(idGroup['takepoint'] > 150.0), 'takepoint']=150.0

idGroup.head(50)
```

```python
# 결과 엑셀파일로 저장
idGroup.to_csv('takepoint.csv')
idGroup.to_csv('takepoint.xlsx')
```
